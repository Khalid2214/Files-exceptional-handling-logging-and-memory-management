Ans1 : The main difference between interpreted and compiled languages lies in how the code is translated into machine language

Compiled Languages:
The source code is translated all at once into machine code by a compiler, producing an executable file (e.g., .exe).

Execution speed: Usually faster because it's already converted into machine code before running.

Interpreted Languages :The source code is read and executed line-by-line by an interpreter at runtime.

Execution speed: Generally slower because it's translated on the fly.

Ans9:the difference between multithreading and multiprocessing?

Feature	Multithreading
Concept	Multiple threads (lightweight processes)

Memory	Shared memory between threads

Best For	I/O-bound tasks (e.g., network, file access)

Speed	Faster context switching

Affected by GIL (only one thread at a time)

Multiprocessing:
Multiple processes (separate memory space)
  	Each process has its own memory
CPU-bound tasks (e.g., heavy computation)
Slower due to process creation overhead
GIL (Python)	Not affected — each process has its own GIL

Ans11 :Memory Management = How Python allocates and reclaims memory while your program runs.

Key Features:
Automatic garbage collection – Python finds and deletes unused objects (via the gc module).

Reference counting – Every object keeps track of how many references point to it.

Private heap space – All Python objects and data structures are stored in a private heap.

Memory pools – Managed internally by Python (via PyMalloc) to optimize performance.

Ans12 :Basic Exception Handling Steps:
try:
- Write code that might raise an exception.

except:
- Handle the error if it occurs.

else: (optional)
- Run if no exceptions were raised in the try block.

finally: (optional)
- Always runs — great for cleanup (like closing files or DB connections)
